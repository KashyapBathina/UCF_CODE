2. Write a function to find the mean, median, and mode of the above created list.

```python
def calculate_stats(random_integers):
    mean = np.mean(random_integers)
    median = np.median(random_integers)
    mode = statistics.mode(random_integers)
    return mean, median, mode

mean, median, mode = calculate_stats(random_integers)
print(f"Mean: {mean}, Median: {median}, Mode: {mode}")
```

### 2. **String Manipulation**

1. Write a function to reverse a string and check if it is a palindrome.
   write you name in example_string_2

```python
example_string_1 = "madam"
example_string_2 = "kashyap" # write your name in the string
#write your code here

def check_if_palindrome(s):
    reversed_s = s[::-1]
    
    if s == reversed_s:
        return True
    else:
        return False

print(check_if_palindrome(example_string_1))
print(check_if_palindrome(example_string_2))
```

2. Create a function that counts the number of vowels in a given string.
   check with above given strings, example_string_1 and example_string_2 and print the output

```python
def count_vowels(s):
    count = 0
    for i in range (len(s)):
        if s[i] in 'aeiouAEIOU':
            count += 1
    return count

print(count_vowels(example_string_1))
print(count_vowels(example_string_2))
```

## **Part 3: Data Visualization with Matplotlib**

### 1. **Plotting Basics**

1. Using the data from the sine and cosine waves, create a subplot that shows both plots individually.
   Hint_1: you can use np.linspace to generate hundred points from 0 to 2π
   Hint_2: use these values of linspace to create sin and cosie values using np.sin() and np.cos()
   Hint_3: Use matplotlib library for and make subplots.

```python
import random
import numpy as np
import statistics
import matplotlib.pyplot as plt


x = np.linspace(0, 2 * np.pi, 100)

sine_values = np.sin(x)
cosine_values = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, sine_values, label='Sine', color='blue')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, cosine_values, label='Cosine', color='red')
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

2. Plot a scatter plot of 100 random points with varying sizes and colors. Include gridlines and labels for the axes. Use numpy and Matplotlib library.

```python
x = np.random.rand(100)
y = np.random.rand(100)

sizes = 1000 * np.random.rand(100)  
colors = np.random.rand(100)        

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.colorbar()  
plt.grid(True)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Points')

plt.show()
```
